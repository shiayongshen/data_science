### Import Dataset

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('googleplaystore.csv')
```

#### Q1

```python
print(df.shape)
```

#### Q2

```python
for i in df:
    if df[i].dtypes == 'float64':
        df[i] = df[i].fillna(df[i].mean())

df = df.dropna()
print(df.shape)
```

#### Q3

```python
df['Size'] = pd.to_numeric(df['Size'].str.rstrip('M'), errors='coerce')
df = df.dropna(subset=["Size"])
df['Installs'] = pd.to_numeric(df['Installs'].str.rstrip('+').str.replace(',', ''), errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df
```

```python
ls = ['Rating','Size','Installs','Reviews']
for i in ls:
    print(df[i].describe())
```

#### Q4

```python
avg_ratings = df.groupby(['Content Rating', 'Genres'])['Rating'].mean().reset_index()

max_avg = avg_ratings['Rating'].max()

top_combinations = avg_ratings[avg_ratings['Rating'] == max_avg]

print("最高平均 Rating:", max_avg)
print("共有這樣的組合數量:", top_combinations.shape[0])
print(top_combinations)
```

#### Q5

```python
avg_ratings = df.groupby(['Category', 'Type'])['Reviews'].mean().reset_index()

max_avg = avg_ratings['Reviews'].max()

top_combinations = avg_ratings[avg_ratings['Reviews'] == max_avg]

print("最高平均 Reviews:", max_avg)
print("共有這樣的組合數量:", top_combinations.shape[0])
print(top_combinations)
```

#### Q6

```python
from scipy import stats

free_ratings = df[df['Type'] == 'Free']['Rating'].dropna()
paid_ratings = df[df['Type'] == 'Paid']['Rating'].dropna()

t_stat, p_value = stats.ttest_ind(free_ratings, paid_ratings, equal_var=False)

print("t:", t_stat)
print("p-value:", p_value)
```

#### Q7

```python
import statsmodels.formula.api as smf
df = df.rename(columns={"Content Rating": "Content_Rating"})

formula = "Rating ~ C(Category) + C(Type) + C(Content_Rating) + C(Genres)"

model = smf.ols(formula, data=df).fit()
print(model.summary())

significant_vars = model.pvalues[model.pvalues < 0.05].drop('Intercept', errors='ignore')
print("\n統計上顯著的變數（p-value < 0.05）:")
print(significant_vars)

print("\nsign var：")
for var in list(significant_vars.index)[:3]:
    print(var)
```

#### Q8

```python
import statsmodels.api as sm
formula = "Rating ~ C(Category) + C(Type) + C(Content_Rating) + C(Genres)"
model = smf.ols(formula, data=df).fit()

anova_results = sm.stats.anova_lm(model, typ=2)
print(anova_results)

significant = anova_results[anova_results['PR(>F)'] < 0.05].sort_values(by='PR(>F)').head(3)
print("\n最顯著的三個變數：")
print(significant)
```
