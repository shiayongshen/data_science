List: Ordered, Mutable, Allow Duplicates

```python
# List
ls = ['s','h','i','a','v']
print(ls[0:3]) #0-2
print(ls[::2]) #each 2
print(ls[::-1]) #reverse
```

```python
ls[2:]
```

```python
ls.insert(0,'hi')
print(ls)
```

```python
ls.remove('hi')
print(ls)
```

```python
ls.count('v')
```
