# F

```python
h_merged = pd.merge(h1, h2, on='id')
print('the primary key for merging is id')
h_merged.to_csv('h_merged.csv', index=False)
```

# G

```python
print(f'Index column names: {h_merged.index.names}') 
print(f'Count of non-missing values: {h_merged.count()}')
print(f'Data types: {h_merged.dtypes}')
```

# H

```python
female_h = h_merged[h_merged['sex'] == 0]
print("Number of observations in female_h:", len(female_h))
```

# I

```python
male_h = h_merged[h_merged['sex'] == 1]
print("Number of observations in male_h:", len(male_h))
```

# J

```python
avg_men = male_h[['cp', 'trestbps', 'chol', 'thalach']].mean()
avg_women = female_h[['cp', 'trestbps', 'chol', 'thalach']].mean()

print(f"男性各指標的平均值：\n{avg_men}")
print(f"女性各指標的平均值：\n{avg_women}")

for col in ['cp', 'trestbps', 'chol', 'thalach']:
    if avg_men[col] > avg_women[col]:
        print(f"{col} - Male: {avg_men[col]:.2f} | Female: {avg_women[col]:.2f}")
```

# K

```python
male_avg_chol = male_h['chol'].mean()
male_chol = male_h[male_h['chol'] < male_avg_chol]

male_chol
```
