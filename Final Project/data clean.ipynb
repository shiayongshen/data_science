##### Find correct answer for product and price

```python
import pandas as pd
df = pd.read_csv('retail_store_sales.csv')
```

```python
df.isnull().sum()
```

```python
#要正確答案的話，需要Item和Price Per Unit都要有值
df = df.dropna(subset=['Item'])

df.isnull().sum()
```

```python
# 查看每個 Category 中的 Item 與其對應的所有 Price Per Unit 值
category_item_price = df.groupby(['Category', 'Item'])['Price Per Unit'].apply(list)

category_item_price_df = category_item_price.reset_index()

category_item_price_df
```

```python
# 檢查每個 Category 和 Item 中的 Price Per Unit 是否有不一樣的值
def check_price_variation(prices):
    return len(set(prices)) > 1

# 新增一列標記是否有不一致的 Price Per Unit
category_item_price_df['Has Variation'] = category_item_price_df['Price Per Unit'].apply(check_price_variation)

# 篩選出有不一致 Price Per Unit 的項目
inconsistent_prices = category_item_price_df[category_item_price_df['Has Variation']]

# 顯示結果
print(inconsistent_prices)
```

```python
# 所以可以使用 mean 來代表其價格
category_item_price = df.groupby(['Category', 'Item'])['Price Per Unit'].mean()

category_item_price_df = category_item_price.reset_index()

category_item_price_df
```

```python
category_item_price_df.duplicated(subset=['Category', 'Item']).sum()
```

```python
# 確保每個 Category 都有25個 Item
category_item_price_df.value_counts(subset=['Category'])
```

```python
# 要確保每個 Category 的 Price Per Unit 都是唯一的
category_item_price_df.duplicated(subset=['Category', 'Price Per Unit']).sum()
```

```python
category_item_price_df.to_csv('category_item_price.csv', index=False)
print("已儲存為 category_item_price.csv")
```

##### Fill up the nan of price

```python
df = pd.read_csv('retail_store_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#將Price Per Unit, Quantity, Total Price 能夠回推計算的補上
#確保至少有兩個欄位有值，然後補上缺失值
def fill_missing_values(row):
    if pd.isnull(row['Price Per Unit']) and not pd.isnull(row['Quantity']) and not pd.isnull(row['Total Spent']):
        row['Price Per Unit'] = row['Total Spent'] / row['Quantity']
    elif pd.isnull(row['Quantity']) and not pd.isnull(row['Price Per Unit']) and not pd.isnull(row['Total Spent']):
        row['Quantity'] = row['Total Spent'] / row['Price Per Unit']
    elif pd.isnull(row['Total Spent']) and not pd.isnull(row['Price Per Unit']) and not pd.isnull(row['Quantity']):
        row['Total Spent'] = row['Price Per Unit'] * row['Quantity']
    return row
df = df.apply(fill_missing_values, axis=1)

print("將Price Per Unit, Quantity, Total Price 能夠回推計算的補上後，資料中仍有空值的欄位")
print(df.isnull().sum())
#可以發現Price Per Unit空值都可以透過Quantity和Total Spent計算出來
```

```python
ans = pd.read_csv('category_item_price.csv')
ans.head()
```

```python
# 若Item有缺少，透過Category 和 Price Per Unit來找出對應的 Item
def get_item(row):
    category = row['Category']
    price_per_unit = row['Price Per Unit']
    item = ans[(ans['Category'] == category) & (ans['Price Per Unit'] == price_per_unit)]['Item'].values
    if len(item) > 0:
        return item[0]
    else:
        return None

df['Item'] = df.apply(get_item, axis=1)
```

```python
df.isnull().sum()
```

```python
#為了確保後續可以針對顧客、線上線下等資訊做insights分析，將資料中的帶有空值的資料以特殊好辨識之值填入，就不輕易刪除
```

```python
#將 Discount Applied 的空值填入 'Unknown'
df['Discount Applied'] = df['Discount Applied'].fillna('Unknown')
#將 Discount Applied 轉換成string格式
df['Discount Applied'] = df['Discount Applied'].astype(str)
```

```python
#將Quantity, Total Price 的空值填入 0
df['Quantity'] = df['Quantity'].fillna(0)
df['Total Spent'] = df['Total Spent'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# 檢查 Price Per Unit, Quantity, Total Price 是否有不合理的資料
df[(df['Total Spent'] < 0) | (df['Price Per Unit'] < 0) | (df['Quantity'] < 0) & (df['Total Spent'] != df['Price Per Unit'] * df['Quantity'])]
```

```python
# 檢查Transaction ID是不是唯一的
df['Transaction ID'].is_unique
```

```python
# 轉換 Transaction Date 的格式
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
# 檢查是否有重複的資料
duplicates = df[df.duplicated()]
duplicates
```

```python
print(f"欄位資料型態分佈：")
print(df.dtypes)
```

```python
print(f"object資料型態分佈：")
for col in df.select_dtypes(include='object').columns:
    print(df[col].apply(type).value_counts())
    print()
```

```python
df.to_csv('retail_store_sales_cleaned.csv', index=False)
print("資料清理完成，已儲存為 retail_store_sales_cleaned.csv")
```

#### Feature engineering

```python
import pandas as pd

df = pd.read_csv('retail_store_sales_cleaned.csv')
df.head()
```

```python
# 拆分日期：從 Transaction Date 提取年、月、日
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df['Year']  = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['Day']   = df['Transaction Date'].dt.day
```

```python
# 取出星期幾（Monday=0 … Sunday=6），若 ≥5 就是週末
df['Is_Weekend'] = df['Transaction Date'].dt.weekday >= 5
```

```python
# 檢查日期是否在假日表裡
import holidays

tw_holidays = holidays.France(years=[2021,2022,2023,2024,2025]).keys()

df['Is_Holiday'] = df['Transaction Date'].dt.date.isin(tw_holidays)
```

```python
# 這邊的假日是指「非工作日」，包含週末與假日
df['Is_NonWorkday'] = (df['Is_Weekend'] | df['Is_Holiday'])
```

```python
# 顧客距離上一次購買天數
df = df.sort_values(['Customer ID', 'Transaction Date'])
df['Prev_Date_Cust'] = df.groupby('Customer ID')['Transaction Date'].shift(1)
df['Recency_Cust'] = (df['Transaction Date'] - df['Prev_Date_Cust']).dt.days
df['Recency_Cust'] = df['Recency_Cust'].fillna(0).astype(int)
```

```python
df = pd.get_dummies(
    df,
    columns=['Payment Method', 'Location', 'Discount Applied'],
    prefix=['PM','Loc','Disc'],
    drop_first=True,
    dtype=int
)
```

```python
df.to_csv('retail_store_sales_cleaned_feature_engineering.csv', index=False)
```
